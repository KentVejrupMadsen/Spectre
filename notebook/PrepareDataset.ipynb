```python
from os.path import (
    isfile, 
    isdir, 
    join
)

from os import (
    listdir, 
    makedirs
)

import random

from shutil \
    import (
        move
)

from math \
    import (
        floor
)
```

```python
original_dataset: str = '/mnt/c/Users/Kentv/Desktop/dataset'
destination_for_dataset: str = '/opt/dataset/tmod'

if isdir(
    destination_for_dataset
):
    %rm {destination_for_dataset} -R

if not isdir(
    destination_for_dataset
):
    makedirs(
        destination_for_dataset
    )
```

```python
found_elements: list = list()

if not isdir(
    destination_for_dataset
):
    makedirs(
        destination_for_dataset
    )

destination_for_training_dataset: str = join(
    destination_for_dataset, 
    'training'
)

destination_for_validation_dataset: str = join(
    destination_for_dataset, 
    'validation'
)


original_dataset_image: str = join(
    original_dataset, 
    'images'
)

original_dataset_label: str = join(
    original_dataset, 
    'labels'
)
```

```python
if not isdir(
    destination_for_dataset
):
    makedirs(
        destination_for_dataset
    )


if not isdir(
    destination_for_training_dataset
):
    makedirs(
        destination_for_training_dataset
    )
```

```python
destination_for_training_images: str = join(
    destination_for_training_dataset, 
    'images'
)

destination_for_training_labels: str = join(
    destination_for_training_dataset, 
    'labels'
)

if not isdir(
    destination_for_training_images
):
    makedirs(
        destination_for_training_images
    )

if not isdir(
    destination_for_training_labels
):
    makedirs(
        destination_for_training_labels
    )
```

```python
if not isdir(
    destination_for_validation_dataset
):
    makedirs(
        destination_for_validation_dataset
    )

destination_for_validation_images: str = join(
    destination_for_validation_dataset, 
    'images'
)

destination_for_validation_labels: str = join(
    destination_for_validation_dataset, 
    'labels'
)

if not isdir(
    destination_for_validation_images
):
    makedirs(
        destination_for_validation_images
    )

if not isdir(
    destination_for_validation_labels
):
    makedirs(
        destination_for_validation_labels
    )
```

```python
Labels: list = list()

def set_labels(
    value: list
) -> None:
    global Labels
    Labels = value

def is_labels_none() -> bool:
    global Labels
    return Labels is None

def get_labels() -> list:
    global Labels
    return Labels
```

```python
def retrieve_label(
    fullpath: str
) -> str | None:
    from os.path \
        import basename
    global original_dataset_label

    filename_for_var = basename(
        fullpath
    ).lower()

    extensions_for_var_file = filename_for_var.split('.')
    size_of_var_extension: int = len(extensions_for_var_file)

    if not is_labels_none():
        set_labels(
            listdir(
                original_dataset_label
            )
        )

        for idx in range(
            len(
                get_labels()
            )
        ):
            get_labels()[idx] = join(
                original_dataset_label, 
                get_labels()[idx]
            )

    found: str | None = None

    for idx in range(
        len(
            get_labels()
        )
    ):
        label_fullpath: str = get_labels()[idx]
        label_file_name: str = basename(label_fullpath)
        label_file_name = label_file_name.lower()
        
        label_extension: list = label_file_name.split('.')

        if len(label_extension) == size_of_var_extension:
            is_the_same: bool = True

            for idx in range(
                len(
                    label_extension
                ) - 1
            ):
                if not(
                    label_extension[idx] == extensions_for_var_file[idx]
                ):
                    is_the_same = False
            
            if is_the_same:
                found = label_fullpath
                return found

    return found
```

```python
for path in listdir(
    original_dataset_image
):
    full_path_for_image: str = join(
        original_dataset_image, 
        path
    )

    found_elements.append(
        {
            'image': full_path_for_image,
            'label': retrieve_label(
                full_path_for_image
            )
        }
    )
```

```python
used_for_validation: float = 0.20

calculated_number_for_split: int = floor(
    len(
        found_elements
    )
    *
    used_for_validation
)
```

```python
for i in range(
    calculated_number_for_split
):
    selected_number = random.randint(
        0, 
        len(
            found_elements
        ) - 1 # Last index number
    )
    
    element = found_elements.pop(
        selected_number
    )
    
    move(
        element['image'], 
        destination_for_validation_images
    )

    move(
        element['label'], 
        destination_for_validation_labels
    )
```

```python
for idx in range(
    len(
        found_elements
    )
):
    element = found_elements[idx]

    move(
        element['image'], 
        destination_for_training_images
    )

    move(
        element['label'], 
        destination_for_training_labels
    )
```
