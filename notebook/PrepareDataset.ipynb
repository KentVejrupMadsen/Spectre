```python
from os.path import (
    isfile, 
    isdir, 
    join
)

from os import (
    listdir, 
    makedirs
)

import random
from shutil import move
from math import floor
```

```python
found_elements: list = list()

original_dataset: str = '/mnt/c/Users/Kentv/Desktop/dataset'
destination_for_dataset: str = '/mnt/c/Users/Kentv/Desktop/TrainingUltralytics/.dataset'

if not isdir(
    destination_for_dataset
):
    makedirs(
        destination_for_dataset
    )

destination_for_training_dataset: str = join(
    destination_for_dataset, 
    'training'
)

destination_for_validation_dataset: str = join(
    destination_for_dataset, 
    'validation'
)


original_dataset_image: str = join(
    original_dataset, 
    'images'
)

original_dataset_label: str = join(
    original_dataset, 
    'labels'
)
```

```python
if not isdir(
    destination_for_dataset
):
    makedirs(
        destination_for_dataset
    )


if not isdir(
    destination_for_training_dataset
):
    makedirs(
        destination_for_training_dataset
    )
```

```python
destination_for_training_images: str = join(
    destination_for_training_dataset, 
    'images'
)

destination_for_training_labels: str = join(
    destination_for_training_dataset, 
    'labels'
)

if not isdir(
    destination_for_training_images
):
    makedirs(
        destination_for_training_images
    )

if not isdir(
    destination_for_training_labels
):
    makedirs(
        destination_for_training_labels
    )
```

```python
if not isdir(
    destination_for_validation_dataset
):
    makedirs(
        destination_for_validation_dataset
    )

destination_for_validation_images: str = join(
    destination_for_validation_dataset, 
    'images'
)

destination_for_validation_labels: str = join(
    destination_for_validation_dataset, 
    'labels'
)

if not isdir(
    destination_for_validation_images
):
    makedirs(
        destination_for_validation_images
    )

if not isdir(
    destination_for_validation_labels
):
    makedirs(
        destination_for_validation_labels
    )
```

```python
for path in listdir(
    original_dataset_image
):
    full_path_for_image: str = join(
        original_dataset_image, 
        path
    )

    full_path_for_label: str = join(
        original_dataset_label, 
        path
    )

    if 'jpg' in full_path_for_label or 'JPG' in full_path_for_label:
        full_path_for_label = full_path_for_label.replace(
            '.jpg', 
            '.txt'
        )

        full_path_for_label = full_path_for_label.replace(
            '.JPG', 
            '.txt'
        )
    
    if 'jpeg' in full_path_for_label or 'JPEG' in full_path_for_label:
        full_path_for_label = full_path_for_label.replace(
            '.jpeg', 
            '.txt'
        )

        full_path_for_label = full_path_for_label.replace(
            '.JPEG', 
            '.txt'
        )

    found_elements.append(
        {
            'image': full_path_for_image,
            'label': full_path_for_label
        }
    )
```

```python
used_for_validation: float = 0.30

calculated_number_for_split: int = floor(
    len(
        found_elements
    )
    *
    used_for_validation
)
```

```python
for i in range(
    calculated_number_for_split
):
    selected_number = random.randint(
        0, 
        len(
            found_elements
        ) - 1 # Last index number
    )
    
    element = found_elements.pop(
        selected_number
    )
    
    move(
        element['image'], 
        destination_for_validation_images
    )

    move(
        element['label'], 
        destination_for_validation_labels
    )
```

```python
for idx in range(
    len(
        found_elements
    )
):
    element = found_elements[idx]

    move(
        element['image'], 
        destination_for_training_images
    )

    move(
        element['label'], 
        destination_for_training_labels
    )

del found_elements
```
